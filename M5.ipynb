## Estadística descriptiva

Si bien ya tenemos estadística descriptiva para los datos, detengamonos en calcular lo pedido.

```python
# el método describe por defecto toma solo las columnas cuantitativas
data.describe().iloc[1:,:].round(2).T
```

Para identificar los outliers acudimos al gráfico `boxplot`, para lo cual recomiendo abordarlo de dos formas, o graficar por separado cada una de las variables, o estandarizar las variables y graficar todo junto, usaremos el segundo, para el cual importamos la clase [`sklearn.preprocessing.StandardScaler`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) de la api [`scikit-learn`](https://scikit-learn.org/stable/index.html).

```python
#!pip install sklearn
from sklearn.preprocessing import StandardScaler

standardData = StandardScaler().fit_transform(data.select_dtypes(include=[np.number]))
standardData = pd.DataFrame(standardData, columns=data.select_dtypes(include=[np.number]).columns)
display(standardData.head())
```

```python
plt.figure(figsize=(12,6)) # Tamño de la figura
sns.boxplot(data=standardData, orient='h',
            medianprops={"color": "r", "linewidth": 2})
plt.title("Boxplot variables numéricas estandarizadas")
plt.tight_layout()
plt.grid(axis='x')
plt.show()
```

Con lo anterior, podemos ver que las variables numéricas no poseen outliers, ni son constantes.

## Análisis de correlación

Para el calculo de las correlaciones consideramos la siguiente hacemos uso de las funciones implementadas en `pandas` para luego graficar un heatmap con sus valores con ayuda de `seaborn`.

```python
corr = data.corr(numeric_only=True)

plt.figure(figsize=(12, 12))
sns.heatmap(data=corr, # Información
            annot=True, # Agregar las anotaciones
            fmt=".2f", # formato de las anotaciones
            mask=np.triu(np.ones(shape=corr.shape)), # Máscara
            vmin=-1,
            vmax=1
            )
plt.title("Matriz de correlaciones")
plt.show()
```

## Visualización de datos

### Distribución de variable numérica

Para esto consideramos el método [`Kernel Density Estimator (KDE)`](https://en.wikipedia.org/wiki/Kernel_density_estimation), para estimar la densidad probabilistica de un varible aleatoria. Podemos usar el parámetro `kde` de algunos graficos.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Rendimiento_Anual'], kde=True, bins=20)
plt.title("Distribución de rendimiento")
plt.xlabel("Rendimiento")
plt.ylabel("Frecuencia")
plt.grid(True, axis='y')
plt.show()
```

### Comparación de variables categóricas con cuantitaticas

Podemos ver la comparación entre distintos tipos de variables con boxplot.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='Nivel_Puesto', y='Rendimiento_Anual', fill=False,
            order=['Junior', 'Mid-level', 'Manager', 'Senior', 'Director'])
plt.title("Rendimiento Anual por Nivel de Puesto")
plt.xlabel("Nivel de Puesto")
plt.ylabel("Rendimiento Anual")
plt.grid(True)
plt.show()
```

### Relación entre variables numéricas

Para analizar la relación entre variables cuantitativas podemos recurrir al uso de `scatterplot`, pero si queremos tener un panorama general `pairplot` es más útil.

```python
sns.pairplot(data=data.select_dtypes(include=[np.number]),
             kind='reg',
             diag_kind='kde',
             corner=True)
plt.title("Pairplot variables numéricas")
plt.tight_layout()
plt.show()
```

### Diferencias entre subgrupos

Para diferenciar diferencias entre subgrupos realizamos la segregación respecto a una variable con el parámetro `hue`.

```python
fig, axs = plt.subplots(3, 1,figsize=(10, 12), sharex=True)

sns.scatterplot(data=data, x='Horas_Extra_Mes', y='Tiempo_Promedio_Proyecto_Dias', hue='Satisfaccion_Laboral', ax=axs[0])
axs[0].grid()

sns.scatterplot(data=data, x='Horas_Extra_Mes', y='Proyectos_Completados', hue='Satisfaccion_Laboral', ax=axs[1])
axs[1].grid()

sns.scatterplot(data=data, x='Horas_Extra_Mes', y='Rendimiento_Anual', hue='Satisfaccion_Laboral', ax=axs[2])
axs[2].grid()
```

## Modelo de regresión


Para los modelos de regresión podemos acudir a las librerías de cálculo científico y/o estadístico. Por ejemplo, podemos hacer uso de la clase `sklearn.linear_model.LinearRegression` de la api `scikit-learn`.

Podemos partir viendo el uso de la regresión lineal simple.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error


X_simple = data[['Salario_USD']]
y = data['Rendimiento_Anual']

model_simple = LinearRegression()              # inicializamos la clase
model_simple.fit(X_simple, y)                  # ajustamos un modelo lineal de X_simple con y

y_pred_simple = model_simple.predict(X_simple) # Predicción en función del ajuste anterior

print("Coeficiente:", model_simple.coef_[0])
print("Intercepto:", model_simple.intercept_)
print("R²:", r2_score(y, y_pred_simple))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred_simple)))

# ========== Plot ====================
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Salario_USD', y='Rendimiento_Anual', data=data, color='blue', label='Datos')
plt.plot(data['Salario_USD'], y_pred_simple, color='red', label='Regresión lineal')

plt.title("Regresión Lineal: Salario vs. Rendimiento")
plt.xlabel("Salario (USD)")
plt.ylabel("Rendimiento Anual")
plt.legend()
plt.grid(True)
plt.show()
```

A partir de la regresión lineal podemos ver a partir del `Salario` podemos explicar un 81% de la variabilidad del rendimiento anual, con un error medio de 0.34 unidades.

```python
X_multi = data[['Salario_USD', 'Horas_Extra_Mes', 'Años_Experiencia', 'Horas_Capacitacion_Anual']]

y = data['Rendimiento_Anual']

model_multi = LinearRegression()
model_multi.fit(X_multi, y)

y_pred_multi = model_multi.predict(X_multi)

print("Coeficientes:", model_multi.coef_)
print("Intercepto:", model_multi.intercept_)
print("R²:", r2_score(y, y_pred_multi))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred_multi)))

# Mostrar cada coeficiente con su variable
for var, coef in zip(X_multi.columns, model_multi.coef_):
    print(f"{var}: {coef:.5f}")
```

Con un modelo un poco más sofisticado como el anterior, podemos ver que sube la explicabilidad a un 92%, con una mejora del error hasta un 0.21 unidades.
